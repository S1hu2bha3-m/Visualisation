```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10 , 200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my title")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sizes = 1000* np.random.rand(50)
plt.scatter(x,y , c= colours , s = sizes , alpha = .2)
# alpha is used to intensity of light
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = ['a','b','c','d','e']
```

```python
y = np.random.rand(5)
plt.bar(x,y) # similarly plt.barh(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title(" bar plot")
```

```python
y
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y, '--b')

plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000* np.random.rand(50)
plt.scatter(x,y , c= 'brown', s = sizes , alpha = .5, marker = 'v')
# alpha is used to intensity of light
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x = iris.sepal_length , y = iris.sepal_width)
```

```python
sns.scatterplot(x = iris.sepal_length , y = iris.petal_length)
```

```python
sns.scatterplot(x = iris.petal_width , y = iris.petal_length)
```

```python
sns.displot(iris['sepal_length'])
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.scatterplot(x = tips.total_bill , y= tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = tips.total_bill , y= tips.tip , data=tips , hue = 'smoker')
```

```python
tips.head()
```

```python
sns.relplot(x = tips.total_bill , y= tips.tip , data=tips , style = 'size')
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x = tips.total_bill , y= tips.tip , data=tips , style = 'size', hue = "time")
```

```python
sns.catplot()
```

```python
tips.head()
```

```python
sns.catplot(x = 'day' , y = 'total_bill' , data = tips)
```

```python
sns.jointplot(x = tips.total_bill , y = tips.tip)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```
